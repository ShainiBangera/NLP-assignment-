```python
#word tokenization
from nltk.tokenize import word_tokenize
text = """This morning was very busy! I woke up early, cleaned my room, and prepared breakfast: eggs, toast, and juice.
After eating, I checked my phone; there were many messages, emails, and reminders. I felt a little tired, but also excited for the day.

In the evening, I went for a walk in the park. The weather was cool, calm, and pleasant; birds were singing, children were playing, and people were relaxing.
I sat on a bench, watched the sunset, and thought about my day: it was simple, yet wonderful!"""
tokens = word_tokenize(text)
print(tokens)
```

```python
#removing punctuations
import string
tokens_rem_punc = [word for word in tokens if word not in string.punctuation]
print(tokens_rem_punc)
```

```python
import nltk
nltk.download('stopwords')
```

```python
#stop words removal
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
print(stop_words)
```

```python
#removing stop words from token
stop_words1 = set(stopwords.words('english'))

filter_words = [w for w in tokens_rem_punc if w.lower() not in stop_words1]
print(filter_words)
```

```python
#stemming
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filter_words]
print(stemmed_words)
```

```python
#lemmatization (shortcut)
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word, pos='v') for word in filter_words]
print (lemmatized_words)
```

```python
#lemmatization (automated)
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

lemmatizer1 = WordNetLemmatizer()

def get_pos(word):
  tag = pos_tag([word])[0][1][0].upper()

  tag_dict = {
      "J": wordnet.ADJ,
      "N": wordnet.NOUN,
      "V": wordnet.VERB,
      "R": wordnet.ADV
  }

  return tag_dict.get(tag, wordnet.NOUN)

lemmatized_words1 = [lemmatizer1.lemmatize(word, get_pos(word)) for word in filter_words]
print(lemmatized_words1)
```

```python
#converting into sentence
converted_text = " ".join(lemmatized_words1)
print(converted_text)
```

**FEATURE ENGINEERING**

```python
#Bag of words
from sklearn.feature_extraction.text import CountVectorizer
corpus = [converted_text]

# Initializing CountVectorizer # It converts text into a matrix of word counts
vectorizer = CountVectorizer()

# Fit the vectorizer to the corpus and transform text into numerical form
vectorized_text = vectorizer.fit_transform(corpus)

print(vectorizer.get_feature_names_out())
print(vectorized_text.toarray())
```

```python
#TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
corpus1 = [converted_text]
vectorizer1 = TfidfVectorizer()
tfidf_vectorized_text = vectorizer1.fit_transform(corpus1)
print(vectorizer1.get_feature_names_out())
print(tfidf_vectorized_text.toarray())
```

```python
#Word2Vec
from gensim.models import Word2Vec
sentences = [
    ["this","morning","was","very","busy"],
    ["i","love","cleaning","my","room"],
    ["i","like","preparing","breakfast"],
    ["prepared","breakfast","eggs","toast","juice"],
    ["breakfast","was","delicious"],
    ["breakfast","included","eggs","toast","juice"],
    ["after","breakfast","i","checked","my","phone"],
    ["there","were","many","messages","emails","and","reminders"],
    ["evening","i","went","for","a","walk","in","the","park"],
    ["I","felt","tired","after","a","busy","day"]
]

model = Word2Vec(sentences,
                 vector_size=100,
                 window=3,
                 min_count=1,
                 epochs=1000)
```

```python
print("Vector for morning")
print(model.wv['morning'])

print("\n Words similar to morning")
print(model.wv.most_similar('morning'))
```

```python
print("Vector for breakfast")
print(model.wv['breakfast'])

print("\n Words similar to breakfast")
print(model.wv.most_similar('breakfast'))
```

```python
print("Vector for tired")
print(model.wv['tired'])

print("\n Words similar to tired")
print(model.wv.most_similar('tired'))
```

```python
import numpy as np

def sent_vec(sentence):
  words = sentence.lower().split() #converting the sentence to lower case and splitting it into words

  vectors =[]
  for word in words:
    if word in model.wv: #checks if the word exists in the Word2Vec vocabulary
      vectors.append(model.wv[word])

  return np.mean(vectors,axis=0) #combining all the vectors into one sentence vector

sentence = "this morning was very busy"
vec = sent_vec(sentence)
print("Sentence Vector:")
print(vec)
```

```python
#cosine similarity(pairwise comparison)
from sklearn.metrics.pairwise import cosine_similarity

# Convert sentences into vectors
s1 = sent_vec("this morning was very busy")
s2 = sent_vec("i had breakfast with eggs")
s3 = sent_vec("evening i went for a walk")

# Compute similarities
similarity1 = cosine_similarity([s1], [s2])[0][0]
similarity2 = cosine_similarity([s1], [s3])[0][0]
similarity3 = cosine_similarity([s2], [s3])[0][0]

print("Sentence Similarity:\n")
print("s1 and s2:", similarity1 )
print("s1 and s3:", similarity2)
print("s2 and s3:", similarity3)
```

```python
#comparing all together
vectors = [s1,s2,s3]

similarity= cosine_similarity(vectors)
print("Sentence Similarity:\n")
print(similarity)
```

```python
#text classification(supervised learning))
from sklearn.linear_model import LogisticRegression

# Convert sentences into numerical vectors using Word2Vec
x = [
    sent_vec("i like preparing breakfast"),
    sent_vec("this morning was lovely"),
    sent_vec("breakfast was delicious"),
    sent_vec("i like having breakfast"),

    sent_vec("i hate getting up early"),
    sent_vec("walking is not my thing"),
    sent_vec("this day is terrible"),
    sent_vec("i feel very tired and bad")
]

y = [1,1,1,1,0,0,0,0] # Labels: 1 = Positive, 0 = Negative

# Create and training the model
model_1 = LogisticRegression()
model_1.fit(x,y)

#testing sentence prediction
test  = sent_vec("i like preparing breakfast")
print("Prediction:", model_1.predict([test]))
```

```python
#clustering(unsupervised learning)
from sklearn.cluster import KMeans

# Convert sentences into vectors
x =[
    sent_vec("i got up early in the morning"),
    sent_vec("i like preparing breakfast"),
    sent_vec("breakfast was delicious"),
    sent_vec("i went for a walk in the morning")
]

kmeans = KMeans(n_clusters=3) # Creating KMeans model with 3 clusters
kmeans.fit(x)

print("Cluster:", kmeans.labels_)
```
